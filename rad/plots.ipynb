```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os
import glob, os
```

```python
def _plot_exp(path_to_exp):
    exp = np.load(path_to_exp).item()
    exp_name = list(exp.keys())[0]

    env_step = [x['env_step'] for x in exp[exp_name].values()]
    mean_ep_reward = [x['mean_ep_reward'] for x in exp[exp_name].values()]
    max_ep_reward = [x['max_ep_reward'] for x in exp[exp_name].values()]
    std_ep_reward = [x['std_ep_reward'] for x in exp[exp_name].values()]

    plt.plot(env_step, mean_ep_reward, label='mean_ep_reward')
    plt.plot(env_step, max_ep_reward, label='max_ep_reward')
    
    plt.fill_between(env_step, np.subtract(mean_ep_reward,std_ep_reward), np.add(mean_ep_reward,std_ep_reward), alpha = 0.1)
    plt.title(exp_name)
    plt.xlabel('env step #')
    plt.ylabel('reward')
    plt.legend()
    plt.savefig(f'output/{exp_name}.jpg')
    plt.show()
```

```python
def plot_domain_task_aug_exp(paths, exp_title):
    for path in paths:
        exp = np.load(path, allow_pickle=True).item()
        exp_name = list(exp.keys())[0]
#         print(exp)

        env_step = [x['env_step'] for x in exp[exp_name].values()]
        mean_ep_reward = [x['mean_ep_reward'] for x in exp[exp_name].values()]
        max_ep_reward = [x['max_ep_reward'] for x in exp[exp_name].values()]
        std_ep_reward = [x['std_ep_reward'] for x in exp[exp_name].values()]
        
        data_aug = path.split('/')[1].split('-')[2]
        plt.plot(env_step, mean_ep_reward, label=data_aug)
#         plt.plot(env_step, max_ep_reward, label='max_ep_reward', label=data_aug)

        plt.fill_between(env_step, np.subtract(mean_ep_reward,std_ep_reward), np.add(mean_ep_reward,std_ep_reward), alpha = 0.1)
    plt.title(exp_title)
    plt.xlabel('env step #')
    plt.ylabel('reward')
    plt.legend()
    plt.savefig(f'outputs/{exp_title}.jpg')
    plt.show()
```
